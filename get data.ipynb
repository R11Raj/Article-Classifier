# Get articles from NYTimes

Terror

```python
lead_para=[]
query=[]
title=[]

for i in range (0,100):
    
    params={'page-size':200,'page':i,'api-key':'ETj0M7oo4IrBHbeA3UGIJ3APSmw2XiwH','q':'terror','begin_date':'20111231','source':'The New York Times'}                                           
    response_articlesearch=requests.get(url='https://api.nytimes.com/svc/search/v2/articlesearch.json',params = params)              
    r=response_articlesearch.json()
    
    for j in r['response']['docs']:
        lead_para.append(j['lead_paragraph'])
        query.append('terror')
        title.append(j['headline']['main'])
    time.sleep(6)
    print(i+1)
    break
```

```python
d1=pd.DataFrame(zip(query,title,lead_para),columns=['Query Name','Title','Content'])
d1.to_csv('Train Data/terror.csv')
print(d1)
```

# Get articles from Guardian API

```python
lead_para=[]
query=[]
title=[]
def remove_html_tags(text):
    """Remove html tags from a string"""
    import re
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

for i in range(1,51):
    
    params={'page-size':200,'page':i,'api-key':'de0c2253-4d8d-4774-aef6-bd73f595df9c','q':'cancer','from-date':'1993-12-31','show-fields':'all'}                                           
    response_articlesearch=requests.get(url='https://content.guardianapis.com/search?',params = params) 
    r=response_articlesearch.json()
    #r1=requests.get(url=r['response']['results'][0]['webUrl'],params=para)
    for j in r['response']['results']:
        lead_para.append(remove_html_tags(j['fields']['body']))
        query.append('cancer')
        title.append(j['fields']['headline'])
    
    time.sleep(12)
    print(i)
```

```python
d1=pd.DataFrame(zip(query,title,lead_para),columns=['Query Name','Title','Content'])
data=pd.read_csv('Train Data/cancer.csv')
data=pd.concat([data,d1],axis=0)
data.to_csv('Train Data/cancer.csv')
```
